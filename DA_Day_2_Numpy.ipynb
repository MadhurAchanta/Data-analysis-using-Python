```python
import numpy as np
a=np.arange(1,13).reshape(2,6)#reshape(rows,columns)
print(a)#arrange(start,stop(inclusive),no of equal parts)
```

```python
a=np.linspace(1,12,12).reshape(2,6)#linspace is a bit different from range.Stop is exclusive in range but here it is inclusive
print(a)
```

```python
b=np.array([1,2,3,4,5,6,7,8]).reshape(2,2,2)#reshape(group,row,column)
print(b)
```

```python
#reshape returns the changed shape and resize reflects the change on the array just like sorted(l) and l.sort()
a=np.array([1,2,3,4])
print(a)
a.reshape(2,2)#returns the new array but no is storing it
print(a)
a.resize(2,2)#resize reflects the change to array
print(a)
```

```python
#adding arrays
a=np.array([1,2,3,4])#we cant perform arithmetic operations on two lists
b=np.array([1,2,3,4])
c=a+b
print(c)
```

```python
#arithmetic operations on arrays
a=np.array([1,2,3,4]).reshape(2,2)
b=np.array([5,6,7,8]).reshape(2,2)
c=b-a
d=a*b#one on one multiplication not matrix multiplication
e=a/b#one of one division
f=a@b#multiplication of matrix
g=a.dot(b)#multiplication of matrix through function
print(c)
print("---------------------")
print(d)
print("---------------------")
print(e)
print("---------------------")
print(f)
print("---------------------")
print("Array g:")
print(g)
print("---------------------")
print("maximum of array g:",g.max())
print("minimum of array g:",g.min())
print("maximum of every column in array g:",g.max(axis=0))#axis=0 is column wise. it prints required value in every column
print("minimum of every row in array g :",g.min(axis=1))#axis=1 is row wise. it prints required value in every row
```

```python
#joining arrays!
a=np.array([1,2,3,4]).reshape(2,2)
b=np.array([5,6,7,8]).reshape(2,2)
print(a)
print("--------------")
print(b)
print("After vertical stacking")
print(np.vstack((a,b)))
print("After horizontal stacking")
print(np.hstack((a,b)))
print(np.stack((a,b),axis=0))#it is similar to vstack
print(np.stack((a,b),axis=1))#it is not hstack but joins nth row of 1st and 2nd array then n+1 row of 1st and 2nd array
```

```python
a=np.arange(30).reshape(2,3,5)
print(a)
print("\nafter dstack\n")#dimentionality stack
print(np.dstack(a))#no of rows--->no of groups
#no of columns in every group ---> no of rows in every group where ---> refers becomes
```

```python
a=np.arange(18).reshape(6,3)
print(a)
print("After vertical splitting")
print(np.vsplit(a,3))#vsplit(array,no of parts)
print("After horizontal splitting")
print(np.hsplit(a,3))
print("non sequential vertical splitting")
print(np.vsplit(a,(2,3)))
print("non sequential horizontal splitting")
print(np.hsplit(a,[2]))#[] is used to split array into two non equal parts
#because if we give number there, it will think we are asking it to divide the array into n equal parts
```

```python
A=np.array([17,21,32,56,43,72,69,100,23,101,5,1]).reshape(6,2)
print(A)
print(A.argmax())#returns index of max element in the array
print(A.argmin())#returns index of min element in the array
print(A.argmax(axis=0))#returns the index of max elements in every column
print(A.argmin(axis=0))#returns the index of min elements in every column
print(A.argmax(axis=1))#returns the index of max elements in every row
print(A.argmin(axis=1))#returns the index of min elements in every row
```

Linear algebra and universal functions

```python
a=np.array([6,24,60,12,18])
b=np.array([6,4,10,5,4])
print(np.add(a,b))#same as a+b
print(a+b)
print(np.subtract(a,b))#same as a-b
print(a-b)
print(np.mod(a,b))#same as a%b
print(a%b)
print(np.divide(a,b))#same as a/b
print(a/b)
print(np.divmod(a,b))
print("Sorted array is :",np.sort(a))
```

```python
a=np.array([6,12,34,56,9,10])
print(np.diff(a))#difference between adjacent two elements
```

set functions


```python
a=np.array([6,24,60,12,18])
b=np.array([6,4,10,5,4])
print(np.union1d(a,b))#a union b
print(np.setdiff1d(a,b))#difference
print(np.intersect1d(a,b))#intersection
```

Rounding

```python
a=np.trunc([-3.8547,3.5896])
print(a)
a=np.round([3.58,2.45])
print(a)
a=np.fix([3.5489,-5.489])
print(a)
a=np.around(3.16645,4)
print(a)
print(np.floor(3.45))
print(np.ceil(3.455))
```

More functions

```python
a=np.array([3,4,16,19,10])
print(np.cumsum(a))#cummulative sum
print(np.cumprod(a))#cummulative product
print(np.lcm(2,6))#lcm of 2 numbers
print(np.gcd(2,6))#gcd of 2 numbers
print(np.lcm.reduce(a))#to find lcm of array
print(np.gcd.reduce(a))#to find gcd of array
```

```python
a=np.array([[1,2],[3,4]])
print(np.linalg.inv(a))#inverse of matrix
```
