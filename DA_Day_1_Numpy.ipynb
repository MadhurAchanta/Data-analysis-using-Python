```python
#Relation operators
#Take a number from the user, which is salary. If the salary is more than 50,000 there is a 10% tax deduction. print his final salary
a=int(input("Enter your salary : "))
if(a>=50000):
  print("Salary is : ",a-(a/10))
else:
  print("Salary is : ",a," which is less than 50000 so no tax deduction")
```

```python
#get the salary from the user, if the salary is more than or equal to 50000 apply 10% tax.
#If the salary is less than 50000 apply 5% tax, if the salary is more than 70000 apply 12% tax
#if salary exceeds one lakh print you are promoted.
a=int(input("Enter your salary : "))
if(a<50000):
  print("Salary is : ",a-(a/20))
elif(a<=70000):
  print("Salary is : ",a-(a/10))
elif(a<=100000):
  print("Salary is : ",a-(a*12/100))
else:
  print("You are promoted!")
```
