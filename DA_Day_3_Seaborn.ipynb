```python
import pandas as pd
import seaborn as sns
sb=pd.read_csv("/content/df2.csv")
pl=sns.lineplot(x="Names",y="SEM1",data=sb)
```

```python
sns.relplot(data=sb,x="SEM1",y="Names",col="Placed")#it is relational plotting, here it will plot same values in same graph
#col==> Graphs are separated based on col
```

```python
dfe=pd.read_excel("/content/diabetes.xlsx")
dfe.tail()
#create a relational plot using age as y axis and index as x axis separated by class
dfe["index"]=range(0,768)
sns.relplot(data=dfe,y="age",x="index",col="class")
```

```python
tips=sns.load_dataset("tips")
#hue=creating difference based on a column via colors
sns.relplot(data=tips,x="total_bill",y="tip",hue="sex")
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",hue="smoker")
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",hue="smoker",style="time",palette="bright")
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",hue="size",palette="bright")
```

```python
abc=sns.load_dataset("dowjones")
sns.relplot(data=abc,x="Date",y="Price",kind="line")#default kind is scatter
```

```python
fmri=sns.load_dataset("fmri")
sns.relplot(data=fmri,x="timepoint",y="signal",kind="line")
```

```python
sns.catplot(data=tips,x="total_bill",y="day")
```

```python
sns.catplot(data=tips,x="smoker",y="total_bill",kind="violin")
```

```python
sns.catplot(data=tips,x="time",y="total_bill",kind="violin",hue="smoker")
```

```python
sns.catplot(data=tips,x="time",y="total_bill",kind="box",hue="smoker")
```
